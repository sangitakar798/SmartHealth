```python

vals = df.values
#symptoms = df1['Symptom'].unique()

#for i in range(len(symptoms)):
#    vals[vals == symptoms[i]] = df1[df1['Symptom'] == symptoms[i]]['weight'].values[0]
    
d = pd.DataFrame(vals, columns=cols)

# Replace certain disease names with 0
d = d.replace('dischromic  patches', 0)
d = d.replace('spotting  urination', 0)
d = d.replace('foul smell of urine', 0)

# Assign the result back to df
df3 = d

# View the last 5 rows of the updated DataFrame
df3
```

```python
# Select only the relevant columns
cols = ['Symptom_1', 'Symptom_2', 'Symptom_3', 'Symptom_4', 'Symptom_5', 'Symptom_6', 'Symptom_7']
df3 = df3[['Disease'] + cols]

# Check if any row has all 0 values
print((df3[cols] == 0).all())

df3[cols] = df3[cols].astype(str)


# Encode the symptoms using one-hot encoding
encoder = OneHotEncoder()
encoded = encoder.fit_transform(df3[cols])


# Print some info about the data
print(df3['Disease'].value_counts())
print(df3['Disease'].unique())

# Prepare data for model training
data = encoded.toarray()
labels = df3['Disease'].values
df3
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
import pandas as pd


# Split data into training and test sets

X_train, X_test, y_train, y_test = train_test_split(data, labels, shuffle=True, train_size=0.85)

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)

model3 = RandomForestClassifier()
model3.fit(X_train, y_train)



train_accuracy = model3.score(X_train, y_train)
print('Training accuracy:', train_accuracy)

test_accuracy = model3.score(X_test, y_test)
print('Test accuracy:', test_accuracy)

y_predict = model3.predict(X_test)
print(y_predict)
# Convert arrays to dataframes
X_train_df = pd.DataFrame(X_train)
X_test_df = pd.DataFrame(X_test)
y_train_df = pd.DataFrame(y_train)

# Save as CSV files
X_train_df.to_csv('X_train.csv', index=False)
X_test_df.to_csv('X_test.csv', index=False)
y_train_df.to_csv('y_train.csv', index=False)
```
